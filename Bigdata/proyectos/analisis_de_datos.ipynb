```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
ventas = pd.read_csv('ventas.csv', sep=';')
```

```python
ventas = ventas.dropna()
```

```python
ventas['fecha'] = pd.to_datetime(ventas['fecha'],format='%d/%m/%Y', errors='coerce')
```

```python
ventas_mes = ventas.groupby([ventas['fecha'].dt.to_period('M')])['cantidad'].sum()
```

```python
ventas_mes.plot(figsize=(10,6), title='Ventas Mensuales')
plt.show()
```

```python
productos_mas_vendidos = ventas['producto'].value_counts().head(10)
```

```python
print(productos_mas_vendidos)
```

```python
plt.figure(figsize=(30,6))
plt.bar(productos_mas_vendidos.index, productos_mas_vendidos.values)
plt.title('Productos Más Vendidos')
plt.show()
```
